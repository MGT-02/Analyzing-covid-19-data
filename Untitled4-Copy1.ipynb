# Top 10 Active cases

```python
top_10_countries = data.groupby('Country/Region')['Active'].mean().nlargest(10)
```

fig = plt.figure(figsize=(16,9))

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_countries.values, y=top_10_countries.index, palette='viridis')

plt.xlabel('Death')
plt.ylabel('Country')
plt.title('Top 10 Countries with the Most COVID-19 Deaths')
plt.show()
```

# Analyse Trends

```python
 total_cases = data.groupby('Date')['Confirmed'].sum()
total_deaths = data.groupby('Date')['Deaths'].sum()
total_recoveries = data.groupby('Date')['Recovered'].sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(total_cases.index, total_cases.values, label='Total Cases')
plt.plot(total_deaths.index, total_deaths.values, label='Total Deaths')
plt.plot(total_recoveries.index, total_recoveries.values, label='Total Recoveries')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('COVID-19 Trends Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

# Daily covid-19 Cases 

```python
daily_cases = data.groupby('Date')['Confirmed'].sum().diff().fillna(0)
daily_deaths = data.groupby('Date')['Deaths'].sum().diff().fillna(0)
daily_recoveries = data.groupby('Date')['Recovered'].sum().diff().fillna(0)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(daily_cases.index, daily_cases.values, label='Daily Cases')
plt.plot(daily_deaths.index, daily_deaths.values, label='Daily Deaths')
plt.plot(daily_recoveries.index, daily_recoveries.values, label='Daily Recoveries')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Daily COVID-19 Cases, Deaths, and Recoveries')
plt.legend()
plt.grid(True)
plt.show()
```
